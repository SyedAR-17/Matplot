```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
coffee_sales = pd.read_csv('/content/index.csv')
coffee_sales.head()
```

```python
coffee_sales.groupby('coffee_name').size().plot(kind='barh', color=sns.dark_palette("xkcd:golden", 8))
plt.xlabel('Number of Sales')
plt.ylabel('Coffee Name')
plt.title('Most Sold Coffee Type')
plt.show()
```

```python
cf = coffee_sales.groupby('cash_type').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.xlabel('Number of Sales')
plt.ylabel('Cash Type')
plt.title('Most Used Cash Type')
plt.show()
```

```python
num_coffees = len(coffee_sales['coffee_name'].unique())
explode = [0.01] * num_coffees
```

```python
plt.pie(coffee_sales['coffee_name'].value_counts(), labels=coffee_sales['coffee_name'].unique(), autopct='%1.0f%%', explode = explode)
plt.title('Coffee Sales')
plt.show()
```
