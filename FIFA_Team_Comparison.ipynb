```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/fifa_data.csv')
```

```python
df.head()
```

```python
BAR = df.loc[df.Club == 'FC Barcelona']['Overall']
RM = df.loc[df.Club == 'Real Madrid']['Overall']
JUV = df.loc[df.Club == 'Juventus']['Overall']

plt.boxplot([BAR, RM, JUV], labels=['FC Barcelona', 'Real Madrid', 'Juventus'], patch_artist= False, boxprops = dict(linestyle='-', linewidth=2, color='darkgoldenrod'), showcaps = True, showmeans = True, medianprops = dict(linestyle='--', linewidth=1.5, color='black'))
plt.title('Comparison of Overall Rating for Different Clubs')

plt.ylabel('Overall Rating')
plt.show()
```
