```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas_prices = pd.read_csv('/content/gas_prices.csv')
```

```python
gas_prices.head()
```

```python
plt.plot(gas_prices.Year, gas_prices.USA, 'b.-')
plt.plot(gas_prices.Year, gas_prices.Canada,'r.-')
plt.plot(gas_prices.Year, gas_prices['South Korea'], 'g.-')
plt.legend(['USA', 'Canada', 'South Korea'])
plt.xticks(gas_prices.Year[::2])
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.title('Gas Prices Over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize':10})
plt.axvline(x=2000, color='black', linestyle = 'dashed',label='axvline, - full height')
plt.show()
```
